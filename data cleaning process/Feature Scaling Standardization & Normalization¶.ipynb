# Feature Scaling
Standardization & Normalization¶

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df2=df[['survived','pclass','age','fare',]]
df2.head()
```

```python
df3=df2.fillna(df.mean())
df3.isnull().sum()
```

```python
X=df3.drop('survived',axis=1)
y=df3['survived']
print('X shape:',X.shape)
print('y shape:',y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)
print('X_train shape:',X_train.shape)
print('y_train shape:',y_train.shape)
print('X_test shape:',X_test.shape)
print('y_test shape:',y_test.shape)
```

```python
sc=StandardScaler()
sc.fit(X_train)
```

```python
sc.mean_
```

```python
sc.scale_
```

```python
X_train.describe()
```

```python
X_train_sc=sc.transform(X_train)
X_test_sc=sc.transform(X_test)
X_train_sc
```

```python
X_train_sc=pd.DataFrame(X_train_sc,columns=['pclass','age','fare'])
X_test_sc=pd.DataFrame(X_test_sc,columns=['pclass','age','fare'])
X_train_sc
```

```python
X_train_sc.describe().round(2)
```

```python
mmc=MinMaxScaler()
X_train_mmc=mmc.fit_transform(X_train)
X_test_mmc=mmc.fit_transform(X_test)
```

```python
X_train_mmc=pd.DataFrame(X_train_mmc,columns=['pclass','age','fare'])
X_test_mmc=pd.DataFrame(X_test_mmc,columns=['pclass','age','fare'])
X_train_mmc
```

```python
X_train_mmc.describe().round(2)
```
