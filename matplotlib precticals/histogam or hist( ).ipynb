```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ML_student_agr=np.random.randint(15,45,(100))
PY_student_agr=np.random.randint(15,40,(100))
```

```python
print(ML_student_agr,PY_student_agr)
```

```python
plt.hist(ML_student_agr)
```

```python
plt.hist(ML_student_agr,bins=[15,20,25,30,35,40,45],rwidth=0.8)
plt.xlabel('age of student')
plt.ylabel('catagory age of student')
plt.title('student')
plt.show()
```

```python
plt.hist(ML_student_agr,bins=[15,20,25,30,35,40,45],rwidth=0.8,orientation='horizontal',color='g')
plt.xlabel('age of student')
plt.ylabel('catagory age of student')
plt.title('student')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.hist(ML_student_agr,bins=[15,20,25,30,35,40,45],rwidth=0.8,orientation='vertical',color='g',label='ML_student_agr')
plt.hist(PY_student_agr,bins=[15,20,25,30,35,40,45],rwidth=0.8,orientation='vertical',color='y',label='PY_student_agr')
plt.xlabel('age of student')
plt.ylabel('catagory age of student')
plt.title('student')
plt.legend()
plt.grid(color='black')
plt.show()
```
