```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students,width=0.5,align='edge',bottom=0.6,color='g',edgecolor='b',linewidth=4,ecolor='r',linestyle='--')
style.use('ggplot')
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.5,align='edge',bottom=0.6,color='g',edgecolor='b',linewidth=4,ecolor='r',linestyle='--',alpha=0.9)

plt.show()
```

```python
plt.figure(figsize=(16,9))
width=0.5
index_classes=np.arange(len(classes))
plt.bar(index_classes,class1_students,width,align='edge',color='g',edgecolor='b',linewidth=4,ecolor='r',linestyle='--',alpha=0.9)
plt.bar(index_classes+width,class2_students,width  ,align='edge',color='k',edgecolor='b',linewidth=4,ecolor='r',linestyle='--',alpha=0.9)
plt.bar(index_classes+width+width,class3_students,width  ,align='edge',color='y',edgecolor='b',linewidth=4,ecolor='r',linestyle='--',alpha=0.9)

plt.title('age index',fontsize=18)
plt.xlabel('classes',fontsize=18)
plt.ylabel('AGE',fontsize=18)
plt.xticks(index_classes+width,classes,rotation=20)
plt.show()
```
