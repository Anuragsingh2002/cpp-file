```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:\\csv file\\bengaluru_house_prices.csv')
df.shape
```

```python
df=pd.read_csv('D:\\csv file\\bengaluru_house_prices.csv',nrows=1000)
df.shape
df
```

```python
x=df['price']
y=df['balcony']
plt.scatter(x,y)
plt.xlabel('price of flat',fontsize=18)
plt.ylabel('balcony',fontsize=18)
plt.title('bengaluru_house_prices',fontsize=18)
```

```python
x=df['price']
y=df['balcony']
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r',s=10,edgecolors='blue',marker='*',alpha=0.5,linewidth=10)
plt.xlabel('price of flat',fontsize=18)
plt.ylabel('balcony',fontsize=18)
plt.title('bengaluru_house_prices',fontsize=18)
plt.figure(figsize=(16,9))
```

```python
x=df['price']
y=df['balcony']
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r',s=10,edgecolors='blue',marker='*',alpha=0.5,linewidth=10)
plt.scatter(x,df['total_sqft'],c='k',s=10,edgecolors='blue',marker='<',alpha=0.5,linewidth=10)

plt.xlabel('price of flat',fontsize=18)
plt.ylabel('balcony',fontsize=18)
plt.title('bengaluru_house_prices',fontsize=18)
```
