```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.pie([1])
plt.show()
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
plt.pie(class1_students)
plt.show()
```

```python
class1_students = [45, 15, 35, 25, 30]
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
plt.figure(figsize=(16,9))
class1_students = [45, 15, 35, 25, 30]
explode=[0.1,0,0,0.3,0.1]
colors=['r','y','g','k','m']
wedgeprops={'linewidth':1,'width':1,'edgecolor':'k'}
plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct='%.2f%%',shadow=True,radius=1.3,textprops={'fontsize':17},
       startangle=90,wedgeprops=wedgeprops,center=(2,3),frame=True,rotatelabels=True,labeldistance=1.7)
plt.legend(loc=1)
plt.show()
```

```python
class1_students = [45, 15, 35, 25, 30]

explode=[0.1,0,0,0.3,0.1]
colors=['r','y','g','k','m']

plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct='%.2f%%',shadow=True,radius=2,textprops={'fontsize':17},
       startangle=90)
plt.legend(loc=1)
plt.show()
```
