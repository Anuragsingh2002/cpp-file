```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\csv file\\class.csv')
df
```

```python
df.pivot_table(index='class')
```

```python
df.pivot_table(index='class',columns='Student id')
```

```python
df.pivot_table(index='class',columns='Student id',aggfunc='sum',margins=True)
```

```python
df.pivot_table(index='class',columns='Student id',aggfunc='sum',margins=True,fill_value=0)
```

```python
df.pivot_table(index='class',columns='Student id',aggfunc='min',margins=True,fill_value=0)
```

```python
df.pivot_table(index='class',columns='Student id',aggfunc='median',margins=True,fill_value=0)
```
